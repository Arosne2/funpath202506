```python
# Add a 'frozen_set' column to each DataFrame and set it as the index
for i, df in enumerate(dfs):
    df['frozen_set'] = df.apply(lambda row: frozenset([row[0], row[1]]), axis=1)
    df.set_index('frozen_set', inplace=True)
    df.drop(columns=[0, 1], inplace=True)  # Drop the first two columns since we're using the index
    df.rename(columns={2: f'Column3_File{i+1}'}, inplace=True)  # Rename the third column
```

```python
#see how many top pairs have overlap
overlap_counts = {}

for i, df1 in enumerate(dfs):
    set1 = set(df1.index)  # Get frozen_set index as a set
    for j, df2 in enumerate(dfs):
        if i < j:  # Avoid duplicate comparisons and self-comparison
            set2 = set(df2.index)
            overlap = set1 & set2  # Intersection of frozen_set indices
            overlap_counts[f'File{i+1} vs File{j+1}'] = len(overlap)

# Display results
for comparison, count in overlap_counts.items():
    print(f"{comparison}: {count} overlapping frozen sets")
```

```python
# see how many top pairs have overlap
overlap_counts = {}
overlap_details = {}  # Store the actual overlapping items
file_names = [df.name if hasattr(df, 'name') else f'File{i+1}' for i, df in enumerate(dfs)]

for i, df1 in enumerate(dfs):
    set1 = set(df1.index)  # Get frozen_set index as a set
    for j, df2 in enumerate(dfs):
        if i < j:  # Avoid duplicate comparisons and self-comparison
            set2 = set(df2.index)
            overlap = set1 & set2  # Intersection of frozen_set indices
            comparison_key = f'{file_names[i]} vs {file_names[j]}'
            overlap_counts[comparison_key] = len(overlap)
            overlap_details[comparison_key] = list(overlap)  # Store the actual overlapping items

# Display results
for comparison, count in overlap_counts.items():
    print(f"{comparison}: {count} overlapping frozen sets")
    if count > 0:  # Only show details if there are overlaps
        print(f"  Common hits: {overlap_details[comparison][:10]}")  # Show first 10
        if count > 10:
            print(f"  ... and {count - 10} more")
    print()
```
